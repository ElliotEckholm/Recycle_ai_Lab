```python
class_name = "trash_wrappers"
```

```python
folder = class_name
file = 'download.csv'
```

```python
path = Path('data/')
dest = path/class_name
dest.mkdir(parents=True, exist_ok=True)
```

```python
dest.ls()
```

## Download Images

```python
# download_images(path/file, dest, max_pics=200)
download_images(path/file, dest, max_pics=400, max_workers=0)
```

#### Remove any images that are corrupted

```python
verify_images(path/class_name, delete=True, max_size=500)
```

## Remove Download.csv file

```python
os.remove(path/file)
```

## View Data

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```
